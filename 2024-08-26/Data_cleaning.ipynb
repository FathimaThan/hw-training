## Image URL

```python
import pandas as pd
import re

def url_qa(df, column_name):
    """
    Perform QA checks on a specified URL column in the DataFrame.

    Parameters:
    - df: DataFrame containing the URL column.
    - column_name: Name of the column to check.

    Returns:
    - A dictionary with DataFrames containing results of the QA checks.
    """
    # Ensure the column exists
    if column_name not in df.columns:
        raise ValueError(f"Column '{column_name}' does not exist in the DataFrame.")
    
    # Convert non-string values to empty strings
    df[column_name] = df[column_name].astype(str)
    
    # Regular expression for validating URLs
    url_regex = re.compile(r'^(https?|ftp)://[^\s/$.?#].[^\s]*$', re.IGNORECASE)
    
    # Function to validate URL format
    def is_valid_url(url):
        return bool(url_regex.match(url))
    
    # Add validation columns
    df[f'{column_name}_valid'] = df[column_name].apply(lambda x: is_valid_url(x) if pd.notna(x) else False)
    df[f'{column_name}_missing_protocol'] = ~df[column_name].str.contains('^https?://', na=False)
    df[f'{column_name}_invalid_chars'] = df[column_name].str.contains(r'[^a-zA-Z0-9:/._\-~]', regex=True, na=False)
    df[f'{column_name}_long'] = df[column_name].str.len() > 2048  # Example length limit
    
    # Identify issues
    missing_protocols = df[df[f'{column_name}_missing_protocol']][[column_name]]
    invalid_chars = df[df[f'{column_name}_invalid_chars']][[column_name]]
    long_urls = df[df[f'{column_name}_long']][[column_name]]
    invalid_urls = df[~df[f'{column_name}_valid']][[column_name]]
    
    # Collect results
    results = {
        'Missing Protocols': missing_protocols,
        'Invalid Characters': invalid_chars,
        'Long URLs': long_urls,
        'Invalid URLs': invalid_urls
    }
    
    return results
```

```python
image_url_qa_results = url_qa(df, 'image_url')

print("\nImage URL QA Results:")
for key, result_df in image_url_qa_results.items():
    print(f"\n{key}:")
    # Exclude rows where 'image_url' is NaN
    print(result_df.dropna(subset=['image_url']))
```

```python
# Regular expression for validating an Email
email_regex = re.compile(r'^[\w\.-]+@[\w\.-]+\.\w+$')

# Function to validate email format
def is_valid_email(email):
    # Convert non-string values to string and handle exceptions
    if isinstance(email, str):
        return bool(email_regex.match(email))
    return False

# Apply validation function
df['valid_email'] = df['email'].apply(is_valid_email)

# Exclude rows with NaN values in 'email' column
df_filtered = df[df['email'].notna()]

# Filter invalid emails, excluding NaN values
invalid_emails = df_filtered[~df_filtered['valid_email']]

print("Invalid email formats:")
print(invalid_emails[['email']])  # Adjust as needed to show relevant columns
```

```python
# Extract domains from email addresses
df['domain'] = df['email'].str.split('@').str[1]

# Count occurrences of each domain
domain_counts = df['domain'].value_counts()

print("Email domain counts:")
print(domain_counts)
```

Email domains such as 'homerealestate.com' and 'HomeRealEstate.com' are treated as different due to case differences, although they refer to the same domain.

```python
# Convert non-string values to empty strings
df['email'] = df['email'].astype(str)

# Identify emails with common mistakes
# Check for missing '@' symbol and invalid characters
common_mistakes = df[
    ~df['email'].str.contains('@', na=False) |  # No '@' symbol
    df['email'].str.contains(r'[^@\w\.-]', regex=True, na=False)  # Invalid characters
]

print("Emails with common mistakes:")
common_mistakes
```
